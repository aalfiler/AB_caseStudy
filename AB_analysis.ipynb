```python
# examine income brackets
def colincometointeger(income):
    if pd.isnull(income):
        return None
    else:
        splitincome = income.split()[0]
        if splitincome == "Prefer":
            return None
        else:
            splitincome = int(splitincome.replace("$", "").replace(",", ""))
            return splitincome

col = 'total_household_income'
new_col = "int_income"
thanks[new_col] = thanks[col].apply(colincometointeger)
print("--------------------")
print(thanks[new_col].value_counts(dropna=False))
print("--------------------")
print(thanks[new_col].describe())
```

# 2.) SQL

```python
select
prd.provider_fullname # identify the provider who placed the order
,count(ord.order_name) # unique Salesforce/CORA order name 
,count(DISTINCT [Patient Unique Patient ID]) # unique patient ID
,count(osf.SpecimenDimKey) # unique samples IDs
,Sum(CASE WHEN ord.order_status_type IN('Completed', 'Active') then 1 else 0 END) # identify orders that have completed sample processing
,DATEDIFF(DAY,prd.[Provider Implied or Actual Create Date PT],GETDATE()) # date the order was placed
from OrderDim ord # sort table by institution account, provider, and patient ID
LEFT JOIN OrderFact orf
on ord.OrderDimKey=orf.OrderDimKey
LEFT JOIN AccountDim ad
on orf.AccountDimKey=ad.AccountDimKey
LEFT JOIN ProviderDim prd
on orf.ProviderDimKey=prd.ProviderDimKey
LEFT JOIN PatientDim pad
on orf.PatientDimKey=pad.PatientDimKey
INNER JOIN OrderSpecimenFact osf
on orf.OrderDimKey=osf.OrderDimKey

SELECT * FROM TableA
LEFT OUTER JOIN TableB
ON TableA.name = TableB.name

WHERE (ord.order_icd_group='ALL' or order_icd10_codes LIKE '%C90.01%') # identify disease diagnosis per ICD 10 codes
and ord.order_category_type='Diagnostic' # only care about clinical orders, not research samples
GROUP BY prd.provider_fullname, DATEDIFF(DAY,prd.[Provider Implied or Actual Create Date PT], GETDATE())
ORDER BY DATEDIFF(DAY,prd.[Provider Implied or Actual Create Date PT],GETDATE())
```
